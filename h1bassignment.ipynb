```python
import pandas as pd
import re

# Suponiendo que tienes un DataFrame llamado 'df' y una columna de texto llamada 'text_column'
# Vamos a limpiar esa columna.
full_dataset['WORKSITE_CITY'] = full_dataset['WORKSITE_CITY'].astype(str)
full_dataset['EMPLOYER_NAME'] = full_dataset['EMPLOYER_NAME'].astype(str)

# Definir una funciÃ³n para limpiar el texto
def clean_text(text):
    # Quitar espacios al inicio y al final
    text = text.strip()
    # Quitar caracteres especiales usando expresiones regulares, manteniendo los espacios entre palabras
    text_clean = re.sub(r'[^\w\s]', '', text)
    # Convertir el texto a mayÃºsculas
    text_clean = text_clean.upper()
    return text_clean

# Aplicar la funciÃ³n de limpieza a la columna de texto
full_dataset['EMPLOYER_NAME'] = full_dataset['EMPLOYER_NAME'].apply(clean_text)
full_dataset['JOB_TITLE'] = full_dataset['JOB_TITLE'].apply(clean_text)
full_dataset['SOC_TITLE'] = full_dataset['SOC_TITLE'].apply(clean_text)
full_dataset['WORKSITE_CITY'] = full_dataset['WORKSITE_CITY'].apply(clean_text)

# Mostrar los resultados

print (len(full_dataset))

full_dataset.head()
```

```python
# FunciÃ³n para limpiar y convertir salarios
def clean_and_convert(column):
    return column.replace('[\$,]', '', regex=True).replace(' ', '').astype(float)

# Aplicar la limpieza a las columnas de salario
h1bvisa['PREVAILING_WAGE'] = clean_and_convert(h1bvisa['PREVAILING_WAGE'])
h1bvisa['WAGE_RATE_OF_PAY_FROM'] = clean_and_convert(h1bvisa['WAGE_RATE_OF_PAY_FROM'])

# Diccionario de factores de conversiÃ³n
conversion_factors = {
    'Hour': 40 * 52,
    'Week': 52,
    'Bi-Weekly': 26,
    'Month': 12,
    'Year': 1
}

# FunciÃ³n vectorizada para calcular los salarios anuales
def calculate_annual_salaries(wage, pay_unit):
    return wage * pay_unit.map(conversion_factors)

# Calcular los salarios anuales
h1bvisa['salary_job'] = calculate_annual_salaries(h1bvisa['PREVAILING_WAGE'], h1bvisa['PW_UNIT_OF_PAY'])
h1bvisa['starting_salary'] = calculate_annual_salaries(h1bvisa['WAGE_RATE_OF_PAY_FROM'], h1bvisa['PW_UNIT_OF_PAY'])

print(h1bvisa[['WAGE_RATE_OF_PAY_FROM', 'PW_UNIT_OF_PAY', 'PREVAILING_WAGE', 'salary_job', 'starting_salary']])
```

```python
# Vamos a utilizar el DataFrame 'h1bvisa' simulado anteriormente para identificar los outliers.

# Calculando la media y la desviaciÃ³n estÃ¡ndar de la columna 'salary_job'
mean_salary = h1bvisa['salary_job'].mean()
std_salary = h1bvisa['salary_job'].std()

# Definiendo un umbral para los outliers basado en 3 desviaciones estÃ¡ndar
outlier_threshold = 3 * std_salary

# Identificando outliers: aquellos valores que estÃ¡n a mÃ¡s de 3 desviaciones estÃ¡ndar de la media
h1bvisa['outlier'] = ((h1bvisa['salary_job'] < (mean_salary - outlier_threshold)) | 
                      (h1bvisa['salary_job'] > (mean_salary + outlier_threshold))).astype(int)

# Contando el nÃºmero de outliers
number_of_outliers = h1bvisa['outlier'].sum()

print(number_of_outliers)
print(h1bvisa[['salary_job', 'outlier']])
```
